```python
# Import required libraries
import random

# Store roll history
roll_history = []

def roll_d6(num_dice=1):
    """
    Roll one or more D6 dice.
    
    Args:
        num_dice: Number of dice to roll (default: 1)
    
    Returns:
        List of roll results
    """
    results = [random.randint(1, 6) for _ in range(num_dice)]
    roll_history.extend(results)
    return results

def display_dice(value):
    """Display a visual representation of a dice face."""
    faces = {
        1: [
            "â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”",
            "â”‚         â”‚",
            "â”‚    â—    â”‚",
            "â”‚         â”‚",
            "â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜"
        ],
        2: [
            "â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”",
            "â”‚  â—      â”‚",
            "â”‚         â”‚",
            "â”‚      â—  â”‚",
            "â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜"
        ],
        3: [
            "â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”",
            "â”‚  â—      â”‚",
            "â”‚    â—    â”‚",
            "â”‚      â—  â”‚",
            "â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜"
        ],
        4: [
            "â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”",
            "â”‚  â—   â—  â”‚",
            "â”‚         â”‚",
            "â”‚  â—   â—  â”‚",
            "â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜"
        ],
        5: [
            "â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”",
            "â”‚  â—   â—  â”‚",
            "â”‚    â—    â”‚",
            "â”‚  â—   â—  â”‚",
            "â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜"
        ],
        6: [
            "â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”",
            "â”‚  â—   â—  â”‚",
            "â”‚  â—   â—  â”‚",
            "â”‚  â—   â—  â”‚",
            "â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜"
        ]
    }
    return faces[value]

def show_stats():
    """Display statistics from all rolls."""
    if not roll_history:
        print("No rolls yet!")
        return
    
    print("\n" + "=" * 35)
    print("         ROLL STATISTICS")
    print("=" * 35)
    print(f"Total rolls:     {len(roll_history)}")
    print(f"Sum of all:      {sum(roll_history)}")
    print(f"Average:         {sum(roll_history) / len(roll_history):.2f}")
    print(f"Highest roll:    {max(roll_history)}")
    print(f"Lowest roll:     {min(roll_history)}")
    print("\nDistribution:")
    for i in range(1, 7):
        count = roll_history.count(i)
        bar = "â–ˆ" * count
        print(f"  {i}: {bar} ({count})")
    print("=" * 35)

def clear_history():
    """Clear the roll history."""
    global roll_history
    roll_history = []
    print("Roll history cleared!")

print("Dice Roller Ready! Run the cells below to roll.")
```

## Roll a Single Dice
Run this cell to roll one D6 dice.

```python
# Roll one dice
result = roll_d6(1)[0]

print("\nğŸ² Rolling one D6...\n")
for line in display_dice(result):
    print(line)
print(f"\nYou rolled: {result}")
```

---
## Roll Multiple Dice
Change the number to roll more dice at once.

```python
# Change this number to roll more dice
num_dice = 3

results = roll_d6(num_dice)

print(f"\nğŸ² Rolling {num_dice} D6 dice...\n")

# Display all dice side by side
all_faces = [display_dice(r) for r in results]
for row in range(5):
    print("  ".join(face[row] for face in all_faces))

print(f"\nResults: {results}")
print(f"Total: {sum(results)}")
```
